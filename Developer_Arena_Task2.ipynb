Number Guessing Game using If-Else Statements

```python
import random

def number_guessing_game():
    secret_number = random.randint(1, 10)
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 10.")

    try:
        guess = int(input("Enter your guess: "))
    except ValueError:
        print("Invalid input. Please enter a whole number.")
        return

    if guess == secret_number:
        print("ðŸŽ‰ Congratulations! You guessed the number correctly!")
    elif guess < secret_number:
        print(f"Too low! The secret number was {secret_number}.")
    else:
        print(f"Too high! The secret number was {secret_number}.")

number_guessing_game()
```

Use For Loop to Print Numbers from 1 to 100

```python
def print_one_to_one_hundred():
    print("Printing numbers from 1 to 100:")
    for i in range(1, 101):
        print(i, end=' ')

print_one_to_one_hundred()
```

Creating a Function that Calculates Area of Rectangle

```python
def calculate_rectangle_area(length, width):
    if length <= 0 or width <= 0:
        return "Error: Length and width must be positive values."

    area = length * width
    return area

area1 = calculate_rectangle_area(10, 5)
area2 = calculate_rectangle_area(7.5, 3)

print(f"\nArea of a 10x5 rectangle: {area1}")
print(f"Area of a 7.5x3 rectangle: {area2}")
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Cannot divide by zero!"
    return x / y

def simple_calculator():
    print("\n--- Simple Calculator ---")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide")

    try:
        choice = input("Enter choice (1/2/3/4): ")
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
    except ValueError:
        print("Invalid number input.")
        return

    if choice == '1':
        print(f"{num1} + {num2} = {add(num1, num2)}")
    elif choice == '2':
        print(f"{num1} - {num2} = {subtract(num1, num2)}")
    elif choice == '3':
        print(f"{num1} * {num2} = {multiply(num1, num2)}")
    elif choice == '4':
        result = divide(num1, num2)
        print(f"{num1} / {num2} = {result}")
    else:
        print("Invalid choice.")

simple_calculator()
```

Real-Life Scenarios: Grading System & Age Categories

```python
def determine_grade(score):
    if score >= 90:
        return "A (Excellent)"
    elif score >= 80:
        return "B (Good)"
    elif score >= 70:
        return "C (Average)"
    elif score >= 60:
        return "D (Needs Improvement)"
    else:
        return "F (Fail)"

print(f"\nScore 95 gets: {determine_grade(95)}")
print(f"Score 72 gets: {determine_grade(72)}")
print(f"Score 55 gets: {determine_grade(55)}")
```

Age Categories

```python
def classify_age(age):
    if age < 0:
        print("Invalid age.")
    elif age <= 12:
        print("A Child")
    elif age > 12 and age <= 19:
        print("A Teenager")
    elif age > 19 and age <= 64:
        print("An Adult")
    else:
        print("A Senior")

print("\nAge 10 is classified as: ", end='')
classify_age(10)
print("Age 18 is classified as: ", end='')
classify_age(18)
print("Age 75 is classified as: ", end='')
classify_age(75)
```
